```python
import requests
import sqlite3
import pandas as pd
import time
```

```python
DB_NAME = "study_and_work_roi.db"
```

```python
def check_database(query=None):
    """Funkcja do sprawdzenia co jest w środku."""
    if query is None:
        query = "SELECT name FROM sqlite_master WHERE type='table';"
    conn = sqlite3.connect(DB_NAME)
    
    df1 = pd.read_sql_query(query, conn)
    conn.close()
    
    return df1
```

```python
query = """
        SELECT * 
        FROM fact_job_postings
        LIMIT 10;
        """
check_database(query)
```
