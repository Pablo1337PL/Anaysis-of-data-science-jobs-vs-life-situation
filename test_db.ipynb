```python
import requests
import sqlite3
import pandas as pd
import time
```

```python
DB_NAME = "study_and_work_roi.db"
```

```python
def check_database(query=None):
    """Funkcja do sprawdzenia co jest w środku."""
    if query is None:
        query = "SELECT name FROM sqlite_master WHERE type='table';"
    conn = sqlite3.connect(DB_NAME)
    
    df1 = pd.read_sql_query(query, conn)
    conn.close()
    
    return df1
```

```python
query = """
        SELECT salary_min, salary_max, url
        FROM fact_job_postings
        WHERE city_name_csv = 'Warsaw'
        LIMIT 10;
        """

with pd.option_context('display.max_columns', None, 'display.max_colwidth', None):
    result = check_database(query)
    print(f"Liczba rekordów: {len(result)}")
    print(result)
```
